Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use plotly express to plot a scatter plot for age 
    and fare columns in the titanic dataset. 

```python
#Ans:
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic
```

```python
import plotly.express as px
fig=px.scatter(titanic,x='age',y='fare')
fig.show()
```

Q3. Using the tips dataset in the plotly library, plot a histogram for x="sex" and y="total_bill" column in the tips dataset.
    Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
    

```python
#Ans:
import plotly.express as px
df=px.data.tips()
fig=px.histogram(df,x='sex',y='total_bill')
fig.show()
```

```python
import plotly.express as px
df=px.data.tips()
fig=px.histogram(df,x='sex',y='total_bill',color='day',pattern_shape='smoker')
fig.show()
```

Q5. What is Displot? Using plotly express, plot a displot.

Ans: The displot figure factory displays a combination of statistical representations of numerical data, such as histogram,
     kernal, density, estimation or normal curve, and rug plot.

```python
from plotly.figure_factory import create_distplot
import numpy as np
x=np.random.randn(100)
hist_data=[x]
group_labels=['displot basic example']
fig=create_distplot(hist_data,group_labels)
fig.show()
```
